```python
from datasets import load_dataset
```

```python
dataset = load_dataset("KANZOO/scrapped_articles")
```

```python
print(dataset)
```

```python
# Function to merge all text into one large dataset
def merge_all_text(dataset):
    combined_text = ""
    for example in dataset['train']:
        combined_text += example['text'] + " "  # Add a space between articles
    return combined_text.strip()

# Merge all text from the dataset
combined_text = merge_all_text(dataset)
```

```python
import re
```

```python
# Regular expression to remove brackets and their contents
cleaned_text = re.sub(r'\(.*?\)', '', combined_text)
```

```python
cleaned_text[:1000]  # Display the first 1000 characters
```

```python
import re

# Function to remove all punctuation
def remove_punctuation(text):
    # Regex pattern to remove all punctuation
    pattern = r'[^\w\s]'  # \w matches word characters, \s matches spaces
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text

# Remove all punctuation
cleaned_text = remove_punctuation(combined_text)

print(cleaned_text[:1000])
```

```python
import stanza

# Download Urdu model
stanza.download('ur')
```

```python
# Load Urdu NLP pipeline
nlp = stanza.Pipeline('ur')
```

```python
# Process the text with Stanza
doc = nlp(cleaned_text)
```

```python
import pandas as pd
# Initialize a list to hold token information
data = []

# Loop through the sentences and tokens in the processed document
for sentence in doc.sentences:
    for token in sentence.tokens:
        # Extract word attributes
        word_info = token.to_dict()[0]  # Stanza stores token info in a list; [0] accesses it
        data.append({
            "id": word_info.get("id"),
            "text": word_info.get("text"),
            "lemma": word_info.get("lemma"),
            "upos": word_info.get("upos"),
            "xpos": word_info.get("xpos"),
            "feats": word_info.get("feats"),
            "head": word_info.get("head"),
            "deprel": word_info.get("deprel"),
            "start_char": word_info.get("start_char"),
            "end_char": word_info.get("end_char"),
            "misc": word_info.get("misc"),
        })

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)
```

```python
df = df.drop(columns=['feats', 'misc'])
```
