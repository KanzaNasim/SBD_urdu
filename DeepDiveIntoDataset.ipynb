```python
from datasets import load_dataset
dataset = load_dataset("KANZOO/scrapped_articles")
```

```python
# Access the 'train' subset
train_dataset = dataset['train']
```

```python
# Select and display the first row
first_row = train_dataset[5]
print(first_row)
```

```python
import stanza

# Download Urdu model
stanza.download('ur')
```

```python
# Load Urdu NLP pipeline
nlp = stanza.Pipeline('ur', processors='tokenize,pos')
```

```python
# Process the text with Stanza
doc = nlp(first_row['text'])

# Save tokens and their XPOS tags in a list
xpos_tags = [
    {'word': word.text, 'xpos': word.xpos}
    for sentence in doc.sentences
    for word in sentence.words
]
```

```python
# Extract unique XPOS tags
unique_xpos_tags = list({tag['xpos'] for tag in xpos_tags})

# Print the unique XPOS tags
print(unique_xpos_tags)
```

```python
# Create a dictionary to store unique tags with corresponding words
unique_tags_with_words = {}

for tag in xpos_tags:
    xpos = tag['xpos']
    word = tag['word']
    if xpos not in unique_tags_with_words:
        unique_tags_with_words[xpos] = set()  # Use a set to avoid duplicate words
    unique_tags_with_words[xpos].add(word)

# Convert sets to lists for easier handling (optional)
unique_tags_with_words = {tag: list(words) for tag, words in unique_tags_with_words.items()}

# Print the result
print(unique_tags_with_words)
```

```python
import json


# Save the output to a JSON file
output_file = 'unique_tags_with_words.json'
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(unique_tags_with_words, f, ensure_ascii=False, indent=4)

print(f"Output saved to {output_file}")
```

```python
# count the number of words
def count_words(text):
    words = text.split()
    return len(words)   
# Count the words in the merged text
word_count = count_words(combined_text)
# Print the total word count        
print("Total Word Count:", word_count)
```

```python
# count special characters
def count_special_characters(text): 
    special_characters = "!@#$%^&*()_+-=[]{}|;:,.<>/?"
    count = 0
    for char in text:   
        if char in special_characters:
            count += 1            
    return count    
# Count the special characters in the merged text
special_char_count = count_special_characters(combined_text)
# Print the total special character count        
print("Total Special Character Count:", special_char_count)
```

```python
#count ngrams
def count_ngrams(text, n):
    words = text.split()    
    ngrams = [words[i:i+n] for i in range(len(words) - n + 1)]
    return len(ngrams)
# Count the n-grams in the merged text
ngram_count = count_ngrams(combined_text, 1)
# Print the total n-gram count        
print("Total N-gram Count:", ngram_count) 
```
